```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
plt.rcParams['axes.facecolor'] = '#F2F2F2'
plt.rcParams['font.family'] = 'CMU Serif'
fig, axes = plt.subplots(figsize=(12,5), nrows=2, ncols=2, sharex=True)

fig.tight_layout()
x = [1, 2, 5, 10, 20]

y00 = [0.963800905, 0.9728506787000001, 0.9760180995000001, 0.9796380090000001, 0.9805429864]
y10 = [1.616289592760181, 2.4624434389140273, 6.160633484162896, 12.227149321266968, 18.228054298642533]

y01 =  [0.5619469027, 0.8343868521000001, 0.8609355247, 0.9121365360000001, 0.9348925411000001]
y11 = [11.38621997471555, 62.09924146649811, 66.23451327433628, 150.12010113780025, 298.5575221238938]

sns.lineplot(x=x, y=y00, ax=axes[0,0], color='#377EB8')
axes[0,0].set_ylabel("Answer Coverage Rate", verticalalignment='baseline', rotation="vertical", labelpad=20)
axes[0,0].set_ylim(0, 1)
axes[0,0].set_title("WikidataSimpleQuestions (max_depth=1)")

axes[1,0].invert_yaxis()
sns.lineplot(x=x, y=y10, ax=axes[1,0], color='#E41A1C')
axes[1,0].set_ylabel("Average Triplets Number", verticalalignment='baseline', rotation="vertical", labelpad=20)
axes[1, 0].set_xticks([1, 2, 5, 10, 20])
axes[1, 0].set_xlabel("Beam Width")

sns.lineplot(x=x, y=y01, ax=axes[0,1], color='#377EB8')
axes[0,1].set_ylim(0, 1)
axes[0,1].set_title("WebQSP (max_depth=2)")


axes[1,1].invert_yaxis()
sns.lineplot(x=x, y=y11, ax=axes[1,1], color='#E41A1C')
axes[1,1].set_xticks([1, 2, 5, 10, 20])
axes[1, 1].set_xlabel("Beam Width")

plt.subplots_adjust(hspace=0, wspace=0.2)
plt.savefig('coverage.svg', format='svg')
```
