#<a id="Introduction"> Introduction</a>
Here under we describe how to process a FTICR2D dataset with NPKv2 software suite.
First are shown the basic commands for making the processing, then how to modify the configuration file according to the experiment.

#<a id="Principle"> Principle</a>
The processing is made with the Python script **processing.py** (in /draft) and a **.mscf** configuration file.  
By default, processing.py searches for a configuration file named **process.mscf**.  
If it does not exist, copy the existing file  **process_eg.mscf** and rename it as **process.mscf**.  
Modify the process.mscf file according to the experiment.   
Then in a Terminal write and execute:   

    mpirun -np nb_of_core python processing.py 
    
by replacing **nb_of_core** by the number of cores you want the processing to be executed on. 

Usually it is preferable to name the configuration file according to the experiment so as to keep the config file attached to the result. In this case, make a config file with the desired name, for example **experim.mscf** and then write and execute in a Terminal:

    mpirun -np nb_of_core python processing.py  experim.mscf

#<a id="Explanations_about_the_configfile"> Explanations about the configfile</a>
There are two main sections indicated between brackets: 
* **[import]**  
* **[processing]**

##[import]

The two main parameters for import are **apex** and **format**.   
**apex** :name of the raw_data file to be processed  
**format** : kind of acquisition preformed (**apex** or **solarix**).   
Becareful, if **infile** is defined in the section [processing], the address given in **apex** is not used.   
**F1_specwidth**, **F2_specwidth** and **highmass** are usually retrieved automatically but for certain experiment it is necessary to provide the right values.  

##[processing]
**infile** : name of the first msh5 file created from the apex file.   
**outfile** : name of the final processed file     
**interfile** : intermediate file, produced after TF in $t_2$.     
**tempdir** : temporary directory   
**sizemultipliers** :  performs zerofilling in F1 and F2.   
**do_F2** : decide if the TF in $t_2$ is performed   
**do_F1** : decide if the TF in $t_1$ is performed    
**do_urqrd** : decide if you use or not the urQRd denoising algorithm.  
**urqrd_rank** : the given integer value is used if do_urqrd is set to True.   
**do_f1demodu: ** used for performing TF in $t_1$ and $t_2$. Always set to True.  
**do_modulus**  : Make the modulus of the processed data. Always set to True.   
**do_rem_ridge** : estimation of the baseline and remove the height found on the whole spectrum.
By default set to False.
